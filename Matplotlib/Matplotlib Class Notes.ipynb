```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day=[1,2,3,4,5,6,7]
Temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(Day,Temp)
plt.show()
```

```python
plt.plot(Day,Temp, color='green',linestyle='dashed',marker='o',linewidth=3)
plt.show()
```

```python
plt.plot(Day,Temp, color='green',linestyle='dotted',marker='o',linewidth=3)
plt.show()
```

```python
plt.xlabel('Day',fontsize=12)
plt.ylabel('Temp',fontsize=12)
plt.title('Weather Info')
plt.plot(Day,Temp, color='blue',linestyle='dotted',marker='o',linewidth=3,alpha=0.5)
plt.show()
```

```python
# legend, Grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day',fontsize=12)
plt.ylabel('Temp',fontsize=12)
plt.title('Weather Info')
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Companies")
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Companies")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Companies")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
plt.show()
```

```python
plt.barh(xpos,revenue)
```

```python
np.random.seed(10)
data_1=np.random.normal(100,10,200)
data_2=np.random.normal(90,20,200)
data_3=np.random.normal(80,30,200)
data_4=np.random.normal(70,40,200)

data=[data_1,data_2,data_3,data_4]
plt.figure(figsize=(10,5))
plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.figure(figsize=(10,7))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0.1,0,0.2,0,0],startangle=45)
plt.legend()
#plt.show()
plt.savefig("piechart.pdf")
```

```python
import os
new_path="E:\AI\librariesforAI\Pandas\Data"
inputFileName="supermarket_sales.csv"

def summaryofData(path,inputFileName):
    pathFileName=path+"/"+inputFileName
    extension = os.path.basename(pathFileName).split(".")[1]
    if extension =="csv":
        df = pd.read_csv(pathFileName)
    elif extension == "xlsx":
        df = pd.read_excel(pathFileName)
    elif extension == "xls":
        df = pd.read_excel(pathFileName)
    elif extension == "json":
        df = pd.read_json(pathFileName)
    else:
        print("Extension not able to found, Please recheck")
        exit()
    print("The Name of Input File : ",inputFileName)
    print("="*60+"\n")
    print("Top 2 rows of the DataFrame :\n\n",df.head(2))
    print("="*60+"\n")
    print("Last 2 rows of the DataFrame :\n\n",df.tail(2))
    print("="*60+"\n")
    print("The shape of the input data is :",df.shape)
    print("="*60+"\n")
    print("Datatype of the input data is :\n\n",df.dtypes)
    print("="*60+"\n")
    print("Basic information about the input data is :\n\n")
    print(df.info())
    print("="*60+"\n")
    print("Describe the numerical Dtypes of input data is :\n\n",df.describe())
    print("="*60+"\n")
    print("Describe the categorical Dtypes of input data is :\n\n",df.describe(include='O'))
    print("="*60+"\n")
    print("Finding the null values if any for the input data :\n\n",df.isnull().sum())
    print("="*60+"\n")
```

```python
summaryofData(new_path,inputFileName)
```
